```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import pandas as pd
```

```python
univ=pd.read_csv('https://raw.githubusercontent.com/slmsshk/DataSet/main/Universities.csv')
```

```python
univ
```

```python
def norm_func(i):
  x=(i-i.min())/(i.max()-i.min())
  return(x)
```

```python
univ_1=norm_func(univ.iloc[:,1:])
```

```python
univ_1
```

```python
# create clusters
hc = AgglomerativeClustering(n_clusters=5, affinity = 'euclidean', linkage = 'complete')
```

```python
hc.fit(univ_1)
```

```python
y_hc=hc.fit_predict(univ_1)
```

```python
y_hc
```

```python
univ_1['h_cluster']=y_hc
```

```python
univ_1
```

```python
u_2=univ_1.sort_values('h_cluster')
```
